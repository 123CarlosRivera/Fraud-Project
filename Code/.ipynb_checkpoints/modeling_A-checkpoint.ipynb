## ```Imports```

````python
# standard libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# modeling libraries
from classifiers_copy1 import classify

# preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
````

## ```Initial Modeling```
---

````python
fraud = pd.read_csv('../SmallBalancedClasses.csv')
````

````python
fraud.head(3)
````

````python
fraud.columns
````

````python
# dummy for object variables
df_fraud = fraud.copy()
df_fraud = pd.get_dummies(fraud, columns=['gender','city','state','category','merchant'])
df_fraud.shape
````

````python
# setting up X/y
X = df_fraud.drop(columns=['trans_date_trans_time','lat', 'long','job','merch_lat', 'merch_long','dob'])
y = df_fraud['is_fraud']

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,stratify=y)

# standard scale the data
ss = StandardScaler()
X_train_ss = ss.fit_transform(X_train)
X_test_ss = ss.transform(X_test)
````

````python
result = classify(X_train_ss,X_test_ss,y_train,y_test)
````
